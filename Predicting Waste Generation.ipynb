##  4 Mathematics behind the Algorithm.
### LSTM (Long Short-Term Memory)
LSTM is a type of recurrent neural network designed to learn long-term dependencies in sequential data. Its mathematics revolve around the concept of "gates" that control information flow:

1. **Forget Gate:**  
   - Equation:
     $$
      f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f) 
     $$
   - It decides what information from the previous cell state to forget.

2. **Input Gate:**  
   - Equation:  
     $$ i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i) $$
   - It determines which new information to add to the cell state.
   - A candidate value is computed as:  
     $$ \tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C) $$

3. **Cell State Update:**  
   - The new cell state is updated as:  
     $$ C_t = f_t \times C_{t-1} + i_t \times \tilde{C}_t $$

4. **Output Gate:**  
   - Equation: 
     $$ o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o) $$
   - The hidden state is then given by:  
     $$ h_t = o_t \times \tanh(C_t) $$

These equations help the LSTM decide what information to remember or forget over time, enabling it to model complex time dependencies.

### XGBoost
XGBoost is a gradient boosting algorithm that builds an ensemble of decision trees. The mathematics here focuses on minimizing a loss function while controlling model complexity:

1. **Objective Function:**  
   - XGBoost minimizes a loss function \( L \) that measures the error between predicted and actual values.  
   - It adds a regularization term \( \Omega \) to penalize overly complex trees:
     $$
     \text{Obj} = \sum_{i=1}^{n} L(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)
     $$

2. **Gradient Boosting:**  
   - At each step, the algorithm computes the gradient (first derivative) and Hessian (second derivative) of the loss function, using them to optimize the split of each decision tree.
   - Each new tree is added to correct the errors made by the previous trees, gradually improving the overall prediction.

### Blended Model
Our blended model simply averages the predictions from the LSTM and XGBoost models. By combining these two approaches, we harness the strength of sequential learning (from LSTM) and robust non-linear pattern recognition (from XGBoost), resulting in improved overall accuracy.

This blend takes advantage of both models' strengths, providing a more reliable prediction than either model alone.

In our journey toward sustainable insights, we begin by gently opening the pages of our CSV files—each a chronicle of waste collected from the bustling heart of a city. We cleanse and harmonize this raw data, resampling it to capture the gentle rhythm of daily totals.

# 5 Python Code used

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import xgboost as xgb

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# =============================================================================
# 1. MERGE ALL CSV FILES INTO A SINGLE DATASET
# =============================================================================
# List of CSV files to merge
files = [
    "boralasgamuwa_2012-2018.csv",
    "dehiwala_2012-2018.csv",
    "homagama_2012-2018.csv",
    "moratuwa_2014-2018.csv",
    "open_source_austin_daily_waste_2003_jan_2021_jul.csv",
    "open_source_ballarat_daily_waste_2000_jul_2015_mar.csv"
]

all_dfs = []
for file in files:
    if not os.path.exists(file):
        print(f"File {file} not found. Skipping.")
        continue
    df_temp = pd.read_csv(file)
    
    # Standardize column names: 'ticket_date' -> 'date', 'net_weight_kg' -> 'waste_amount'
    if 'ticket_date' in df_temp.columns:
        df_temp.rename(columns={'ticket_date': 'date'}, inplace=True)
    if 'net_weight_kg' in df_temp.columns:
        df_temp.rename(columns={'net_weight_kg': 'waste_amount'}, inplace=True)
    
    # Create a 'city' column from the filename (capitalized)
    city_name = file.split('_')[0].capitalize()
    df_temp['city'] = city_name
    all_dfs.append(df_temp)

# Concatenate all DataFrames
df_merged = pd.concat(all_dfs, ignore_index=True)
# Drop rows with missing date or waste_amount
df_merged.dropna(subset=['date', 'waste_amount'], inplace=True)

# Convert 'date' column to datetime and set as index
df_merged['date'] = pd.to_datetime(df_merged['date'])
df_merged.set_index('date', inplace=True)

print("Merged dataset shape:", df_merged.shape)

# =============================================================================
# 2. DATA CLEANING & RESAMPLING
# =============================================================================
# Aggregate data to daily totals (summing waste amounts from all cities)
daily_data = df_merged.resample('D').sum()
daily_data.fillna(0, inplace=True)

# =============================================================================
# 3. FEATURE ENGINEERING
# =============================================================================
# Add cyclical day-of-week features (to capture weekly patterns)
daily_data['day_of_week'] = daily_data.index.dayofweek  # Monday=0, Sunday=6
daily_data['dow_sin'] = np.sin(2 * np.pi * daily_data['day_of_week'] / 7)
daily_data['dow_cos'] = np.cos(2 * np.pi * daily_data['day_of_week'] / 7)
daily_data.drop(columns=['day_of_week'], inplace=True)

# Add cyclical month features (to capture seasonal variations)
daily_data['month'] = daily_data.index.month  # Month as integer (1-12)
daily_data['month_sin'] = np.sin(2 * np.pi * daily_data['month'] / 12)
daily_data['month_cos'] = np.cos(2 * np.pi * daily_data['month'] / 12)
daily_data.drop(columns=['month'], inplace=True)

# Reorder columns: target (waste_amount) first, then cyclical features
daily_data = daily_data[['waste_amount', 'dow_sin', 'dow_cos', 'month_sin', 'month_cos']]

# =============================================================================
# 4. SCALING
# =============================================================================
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(daily_data.values)
df_scaled = pd.DataFrame(scaled_data, index=daily_data.index, columns=daily_data.columns)

# =============================================================================
# 5. CREATE SEQUENCES
# =============================================================================
def create_sequences(data, look_back):
    """
    Creates sequences for time series forecasting.
    X: Sequences of shape (look_back, n_features)
    y: Target value (waste_amount) following the sequence.
    """
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back, :])
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

look_back = 10
X, y = create_sequences(df_scaled.values, look_back)

# =============================================================================
# 6. TRAIN/TEST SPLIT
# =============================================================================
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

# =============================================================================
# 7. BUILD LSTM MODEL
# =============================================================================
n_features = X.shape[2]  # e.g., 5 features: [waste_amount, dow_sin, dow_cos, month_sin, month_cos]
lstm_model = Sequential([
    LSTM(64, activation='tanh', return_sequences=True, input_shape=(look_back, n_features)),
    Dropout(0.2),
    LSTM(64, activation='tanh'),
    Dropout(0.2),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = lstm_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1,
                         callbacks=[early_stop], verbose=1)

lstm_pred = lstm_model.predict(X_test)

# =============================================================================
# 8. BUILD XGBOOST MODEL
# =============================================================================
# Flatten the 3D input for XGBoost (concatenating all timesteps and features)
X_train_flat = X_train.reshape(X_train.shape[0], look_back * n_features)
X_test_flat = X_test.reshape(X_test.shape[0], look_back * n_features)

xgb_model = xgb.XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=5)
xgb_model.fit(X_train_flat, y_train)
xgb_pred = xgb_model.predict(X_test_flat)

# =============================================================================
# 9. BLEND PREDICTIONS
# =============================================================================
blend_pred = 0.5 * lstm_pred.flatten() + 0.5 * xgb_pred

# =============================================================================
# 10. INVERSE SCALING
# =============================================================================
def inverse_transform_targets(y_scaled):
    """
    Inverses the scaling for target values (waste_amount).
    Creates a dummy array with the same shape as the original features,
    replacing the target column with the scaled predictions.
    """
    dummy = np.zeros((len(y_scaled), daily_data.shape[1]))
    dummy[:, 0] = y_scaled  # target is the first column
    inv = scaler.inverse_transform(dummy)
    return inv[:, 0]

lstm_pred_inv = inverse_transform_targets(lstm_pred.flatten())
xgb_pred_inv = inverse_transform_targets(xgb_pred)
blend_pred_inv = inverse_transform_targets(blend_pred)
y_test_inv = inverse_transform_targets(y_test)

# =============================================================================
# 11. EVALUATION
# =============================================================================
lstm_rmse = np.sqrt(mean_squared_error(y_test_inv, lstm_pred_inv))
xgb_rmse = np.sqrt(mean_squared_error(y_test_inv, xgb_pred_inv))
blend_rmse = np.sqrt(mean_squared_error(y_test_inv, blend_pred_inv))

print(f"LSTM RMSE:  {lstm_rmse:.2f}")
print(f"XGBoost RMSE: {xgb_rmse:.2f}")
print(f"Blended RMSE: {blend_rmse:.2f}")

# =============================================================================
# 12. PLOT RESULTS
# =============================================================================
plt.figure(figsize=(10, 6))
plt.plot(y_test_inv, label='Actual', alpha=0.8)
plt.plot(lstm_pred_inv, label='LSTM Prediction', alpha=0.7)
plt.plot(xgb_pred_inv, label='XGBoost Prediction', alpha=0.7)
plt.plot(blend_pred_inv, label='Blended Prediction', alpha=0.7)
plt.title("Daily Solid Waste Prediction (Blended Model)")
plt.xlabel("Time (Days)")
plt.ylabel("Waste Amount (Original Scale)")
plt.legend()
plt.show()
```

While the blended approach slightly improves upon the LSTM model, it does not outperform XGBoost in this instance. These findings imply that, for our current dataset, XGBoost is more effective at capturing the underlying patterns in municipal solid waste data, though further tuning or additional features might improve the blended model's performance.

# 7. Implementation and Interpretation.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import xgboost as xgb

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

### LOAD DATA

```python
# Set the file path for the chosen dataset (single-city example for demonstration)
file_path = "boralasgamuwa_2012-2018.csv"

# Load data from the CSV file
df = pd.read_csv(file_path)

# Standardize column names:
# Rename 'ticket_date' to 'date' and 'net_weight_kg' to 'waste_amount'
df.rename(columns={'ticket_date': 'date', 'net_weight_kg': 'waste_amount'}, inplace=True)

# Parse the 'date' column and set it as the index
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

### 2. DATA CLEANING & RESAMPLING

```python
# Aggregate data to daily frequency by summing waste amount
daily_data = df.resample('D').sum()

# Fill missing days (if any) with 0 waste
daily_data.fillna(0, inplace=True)
```

### 3. FEATURE ENGINEERING

```python
# Add cyclical day-of-week features to capture weekly seasonality
daily_data['day_of_week'] = daily_data.index.dayofweek  # Monday=0, Sunday=6
daily_data['dow_sin'] = np.sin(2 * np.pi * daily_data['day_of_week'] / 7)
daily_data['dow_cos'] = np.cos(2 * np.pi * daily_data['day_of_week'] / 7)
daily_data.drop(columns=['day_of_week'], inplace=True)

# Add cyclical month features to capture seasonal variations
daily_data['month'] = daily_data.index.month  # Month as integer (1-12)
daily_data['month_sin'] = np.sin(2 * np.pi * daily_data['month'] / 12)
daily_data['month_cos'] = np.cos(2 * np.pi * daily_data['month'] / 12)
daily_data.drop(columns=['month'], inplace=True)

# Reorder columns so that the target ('waste_amount') is first, followed by cyclical features
# Final feature order: [waste_amount, dow_sin, dow_cos, month_sin, month_cos]
daily_data = daily_data[['waste_amount', 'dow_sin', 'dow_cos', 'month_sin', 'month_cos']]
```

### 4. SCALING

```python
# Scale all features together using MinMaxScaler (range 0-1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(daily_data.values)

# (Optional) Convert back to a DataFrame for clarity
df_scaled = pd.DataFrame(scaled_data, index=daily_data.index, columns=daily_data.columns)
```

### 5. CREATE SEQUENCES

```python
def create_sequences(data, look_back):
    """
    Creates sequences from the data for time series forecasting.
    X: sequences of shape (look_back, n_features)
    y: target value (waste_amount) following the sequence.
    """
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back, :])  # all features for past 'look_back' days
        y.append(data[i + look_back, 0])      # target: waste_amount (first column)
    return np.array(X), np.array(y)

# Set look-back window (number of past days)
look_back = 10
X, y = create_sequences(df_scaled.values, look_back)
```

### 6. TRAIN/TEST SPLIT

```python
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
```

### 7. BUILD LSTM MODEL

```python
n_features = X.shape[2]  # e.g., now 5 features: [waste_amount, dow_sin, dow_cos, month_sin, month_cos]

lstm_model = Sequential([
    LSTM(64, activation='tanh', return_sequences=True, input_shape=(look_back, n_features)),
    Dropout(0.2),
    LSTM(64, activation='tanh'),
    Dropout(0.2),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the LSTM model
history = lstm_model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=1
)

# Predict on the test set using the LSTM model
lstm_pred = lstm_model.predict(X_test)
```

### 8. BUILD XGBOOST MODEL

```python
# Flatten the 3D input for XGBoost (all timesteps and features)
X_train_flat = X_train.reshape(X_train.shape[0], look_back * n_features)
X_test_flat = X_test.reshape(X_test.shape[0], look_back * n_features)

# Train an XGBoost regressor
xgb_model = xgb.XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=5)
xgb_model.fit(X_train_flat, y_train)

# Predict using the XGBoost model
xgb_pred = xgb_model.predict(X_test_flat)
```

### 9. BLEND PREDICTIONS

```python
# Blend predictions by taking a simple average of LSTM and XGBoost outputs
blend_pred = 0.5 * lstm_pred.flatten() + 0.5 * xgb_pred
```

### 10. INVERSE SCALING

```python
def inverse_transform_targets(y_scaled):
    """
    Inverses the scaling for target values (waste_amount).
    Since the scaler was applied to all features, create a dummy array and
    replace the target column with the scaled predictions.
    """
    dummy = np.zeros((len(y_scaled), daily_data.shape[1]))
    dummy[:, 0] = y_scaled  # target is the first column
    inv = scaler.inverse_transform(dummy)
    return inv[:, 0]

# Inverse transform predictions and true values for interpretability
lstm_pred_inv = inverse_transform_targets(lstm_pred.flatten())
xgb_pred_inv = inverse_transform_targets(xgb_pred)
blend_pred_inv = inverse_transform_targets(blend_pred)
y_test_inv = inverse_transform_targets(y_test)
```

### 11. EVALUATION

```python
lstm_rmse = np.sqrt(mean_squared_error(y_test_inv, lstm_pred_inv))
xgb_rmse = np.sqrt(mean_squared_error(y_test_inv, xgb_pred_inv))
blend_rmse = np.sqrt(mean_squared_error(y_test_inv, blend_pred_inv))

print(f"LSTM RMSE:  {lstm_rmse:.2f}")
print(f"XGBoost RMSE: {xgb_rmse:.2f}")
print(f"Blended RMSE: {blend_rmse:.2f}")
```

### 12. PLOT RESULTS

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test_inv, label='Actual', alpha=0.8)
plt.plot(lstm_pred_inv, label='LSTM Prediction', alpha=0.7)
plt.plot(xgb_pred_inv, label='XGBoost Prediction', alpha=0.7)
plt.plot(blend_pred_inv, label='Blended Prediction', alpha=0.7)
plt.title("Daily Solid Waste Prediction (Blended Model)")
plt.xlabel("Time (Days)")
plt.ylabel("Waste Amount (Original Scale)")
plt.legend()
plt.show()
```
